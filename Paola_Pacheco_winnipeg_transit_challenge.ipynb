````python
# Iterate through rows in df_cleaned to get the coordinates
for index, row in df_cleaned.iterrows():

    # Get the origin address from df_cleaned
    origin_address_to_check = row['origin_address']

    # Get the destination address from df_cleaned
    destination_address_to_check = row['destination_address']

    # Check if the origin address (df_cleaned) is in reference_df
    matching_origin_row = reference_df[reference_df['origin_address'] == origin_address_to_check]

    # Check if the destination address (df_cleaned) is in reference_df
    matching_destination_row = reference_df[reference_df['destination_address'] == destination_address_to_check]

    # If a match is found, update the coordinates in df_cleaned
    if not matching_origin_row.empty:
        df_cleaned.at[index, 'origin_xcoordinate'] = matching_origin_row['origin_xcoordinate'].values[0]
        df_cleaned.at[index, 'origin_ycoordinate'] = matching_origin_row['origin_ycoordinate'].values[0]

    if not matching_destination_row.empty:
        df_cleaned.at[index, 'destination_xcoordinate'] = matching_destination_row['destination_xcoordinate'].values[0]
        df_cleaned.at[index, 'destination_ycoordinate'] = matching_destination_row['destination_ycoordinate'].values[0]
````
